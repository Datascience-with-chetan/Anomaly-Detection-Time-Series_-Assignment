1.  What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.
- Anomaly Detection is the process of identifying unusual data points, patterns, or behaviors that deviate from what is considered normal in a dataset or system. It is used in fields such as cybersecurity, fraud detection, healthcare monitoring, and system reliability to spot errors, threats, and novel events rapidly.

- Types of Anomalies
  - Point Anomalies
    - A point anomaly occurs when an individual data point stands out significantly from the rest of the data set.

    - Example: A sudden and very large credit card transaction compared to usual purchases for that user indicates potential fraud.

    - Example: A single network spike in traffic that is far above the normal range could signal an attack or malfunction.

  - Contextual Anomalies
    - A contextual anomaly is a data point that is only anomalous in a certain context such as time, location, or user activity.

    - Example: High energy consumption at midday in a home when no one is normally present could signal a malfunction or intrusion.

    - Example: An 85°F temperature reading is normal in summer but highly abnormal in winter.

    - Example: Logging into a business system outside regular hours from an unusual location may indicate a security breach.

  - Collective Anomalies
    - Collective anomalies refer to groups or sequences of data points that deviate from expected collective patterns, even when each individual point seems normal.

    - Example: In healthcare, a run of slightly abnormal heartbeats that, taken together, indicate a cardiac event, even though each single beat falls within typical limits.

    - Example: In network security, multiple login attempts across several accounts within a short period may indicate a coordinated attack (such as a brute-force or DDoS attack), even though each single attempt is not anomalous.

    - Example: Consistently late deliveries from a supplier form a pattern that points to a supply chain issue, while individual delays might not stand out alone.

2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
- Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are popular anomaly detection algorithms, each with distinct approaches and ideal scenarios.
- **Isolation Forest**
    - Works by recursively partitioning the data space, randomly selecting features and split values; anomalies are easier to isolate, so they have shorter average path lengths in partitioning trees.

    - Handles large datasets well, scales efficiently to high-dimensional data, and does not require assumptions about data distribution.

    - Suitable for general applications like fraud detection, system failure monitoring, and any scenario with many features or large sample sizes.

    - Requires specifying a contamination parameter (expected proportion of outliers), and can struggle if there are many anomalies or the distinction between normal and abnormal points is subtle.

- Isolation Forest Usecases:

    - Effective for high-dimensional data or very large datasets (e.g., fraud detection, fault detection in server logs).

    - Mixed data types (numerical and categorical) can be handled.

    - When anomalies are expected to be few and different.

- **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**
    - Groups points into clusters based on density; points that do not belong to any cluster (because they're not dense enough) are labeled as noise/anomalies.

    - Effective for datasets with clear cluster structure and low-to-medium dimensions, especially spatial or geographical data or customer segmentation.

    - Can find anomalies in arbitrary-shaped clusters and is robust to noise, but sensitive to parameter settings (eps and min_samples).

    - Does not perform well on very high-dimensional data due to difficulty defining distance and density, and may struggle with clusters of widely varying density.

- DBSCAN Usecases:

    - Best for spatial data or data with naturally occurring clusters of differing shapes (e.g., geospatial anomalies, customer segmentation).

    - Good for situations with arbitrary cluster boundaries and noise.

    - Less effective for high-dimensional data

- **Local Outlier Factor (LOF)**
    - Measures the local density of data points and calculates how much a point differs from its neighbors; finds anomalies that are “isolated” relative to the surrounding region.

    - Excels at detecting local anomalies where groups of normal points have different underlying densities, like sensor failures or fraud in a naturally segmented population.

    - Ideal when anomalies are only out-of-place with respect to their context, not globally unusual, and works best in low-to-medium dimensional data.

    - Sensitive to parameter choices (number of neighbors) and computationally heavier, especially as dataset grows.

- Local Outlier Factor Usecases:

    - Preferred in datasets with clusters of varying density or when local anomalies are the main concern (e.g., intrusion detection, rare event spotting in sensor data).

    - Useful when anomalies deviate with respect to their local neighborhood instead of globally.

    - Requires careful parameter tuning, more computational cost


3.What are the key components of a Time Series? Explain each with one example.
- The key components of a time series are Trend, Seasonality, Cyclicity, and Irregular (Random) Variations. Each component reveals different information about changes and patterns over time.

- Trend
  - Description: The trend shows the long-term movement or direction in the data, either upward, downward, or flat, across a significant time period. It represents the general tendency, not short-term fluctuations.

  - Example: The steady increase in monthly average global temperature over decades due to climate change.

- Seasonality
  - Description: Seasonality refers to regular patterns that repeat at fixed intervals within the data, usually within a year (daily, monthly, quarterly, etc).

  - Example: Retail sales peaking every December due to holiday shopping, and dropping in January.

- Cyclical Variation
  - Description: Cyclical fluctuations are long-term oscillations (more than a year) not fixed to a calendar, often related to the broader economic or business cycles. Unlike seasonality, cycles do not have a set length.

  - Example: Recessions and expansions in GDP that occur every few years.

- Irregular (Random) Variation
  - Description: Irregular or random variation refers to unsystematic, unpredictable changes that do not follow a pattern and arise from unforeseen events.

  - Example: A sudden drop in airline travel following an unexpected global event like a pandemic or natural disaster

4.Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?
- Definition of Stationary in Time Series
A time series is said to be stationary if its statistical properties—such as mean, variance, and autocorrelation—remain constant over time and do not depend on the time at which the series is observed. This means that:

  - The mean of the series is constant (no upward or downward trend).

  - The variance is constant over time (stable fluctuations).

  - The autocovariance between observations depends only on the lag (time difference) between observations, not on the actual time.

- Stationarity is important because many time series forecasting and modeling techniques assume the time series to be stationary for better accuracy and reliability.

- How to Transform a Non-Stationary Series into a Stationary One
- Differencing

  - Compute the difference between consecutive observations.

  - Can be applied multiple times (first difference, second difference) to remove trends and stabilize mean.

- Detrending

  - Remove trends using regression or subtracting a moving average.

  - Linear or nonlinear trends can be modeled and subtracted from the series.

- Seasonal Differencing

  - Difference the series at seasonal lags (e.g., subtract value from the same season in the previous cycle) to remove seasonality.

- Transformation

  - Apply transformations such as logarithm, square root, or Box-Cox to stabilize variance.

- Combination
  
  - Often a combination of differencing, detrending, and transformation is used to achieve stationarity before modeling.





5.Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.
- Autoregressive Model (AR)
  - Structure: AR models the current value of the series as a linear combination of its past values (lags). It is denoted as AR(p), where
p is the number of lag observations included.

  - Application: Suitable for stationary time series where past values have a direct influence. Used in financial stock prices or weather data modeling when the past influences the present.

  - Example: Predict current temperature based on previous few days.

- Moving Average Model (MA)
  - Structure: MA models the current value as a linear combination of past forecast errors (residuals). It is denoted as MA(q), where
q is the number of lagged error terms used.

  - Application: Useful to model series where the error terms influence the present, smoothing out noise.

  - Example: Modeling daily sales where recent shocks impact current demand.

- Autoregressive Integrated Moving Average (ARIMA)
  - Structure: Combines AR and MA with differencing (I) to make the time series stationary. Notation ARIMA(p, d, q) where
d is the order of differencing to remove trend/non-stationarity.

  - Application: Versatile model for non-stationary series with trends but no seasonality. Widely used in various forecasting tasks.

  - Example: Predicting monthly sales with a trend but no repeating seasonal pattern.

- Seasonal ARIMA (SARIMA)
  - Structure: Extends ARIMA to handle seasonality by adding seasonal AR, MA, and differencing terms: SARIMA(p, d, q),(P, D, Q) s where is the seasonal period.

  - Application: Suitable for data with seasonal patterns like monthly retail sales with holiday spikes or weather data showing yearly cycles.

  - Example: Forecasting quarterly revenue with yearly seasonal effects.

- SARIMAX (Seasonal ARIMA with eXogenous variables)
  - Structure: Further extends SARIMA by incorporating exogenous regressors (independent variables) to explain the series along with seasonal ARIMA components.

  - Application: Useful when external drivers influence the time series, like economic indicators affecting sales.

  - Example: Forecasting energy demand using past demand plus temperature and economic activity as external predictors.

```python
#6.Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load dataset from the attached CSV file
df = pd.read_csv('AirPassengers.csv', parse_dates=['Month'], index_col='Month')

# Rename column for clarity
df.rename(columns={'#Passengers': 'Passengers'}, inplace=True)

# Plot original time series
plt.figure(figsize=(12, 4))
plt.plot(df['Passengers'], label='Original Series', color='blue')
plt.title('AirPassengers Time Series')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()

# Decompose the time series (multiplicative model as variance seems to increase over time)
decomposition = seasonal_decompose(df['Passengers'], model='multiplicative', period=12)

# Plot decomposition
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.suptitle('Decomposition of AirPassengers Time Series', fontsize=16)
plt.show()


```

```python
#7. Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Simulate a numerical dataset similar to NYC Taxi Fare parameters (e.g. fare amount and trip distance)
np.random.seed(42)
normal_data = np.random.randn(300, 2) * [10, 5] + [50, 10]  # Normal cluster around (50, 10)
anomalies = np.random.uniform(low=[80, 20], high=[120, 60], size=(20, 2))  # Anomalies far from cluster

# Combine the dataset
data = np.vstack([normal_data, anomalies])
df = pd.DataFrame(data, columns=['Fare_Amount', 'Trip_Distance'])

# Apply Isolation Forest
iso_forest = IsolationForest(contamination=0.06, random_state=42)
df['anomaly'] = iso_forest.fit_predict(df)

# Mark anomalies (label = -1) and normal points (label = 1)
anomalies_df = df[df['anomaly'] == -1]
normal_df = df[df['anomaly'] == 1]

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(normal_df['Fare_Amount'], normal_df['Trip_Distance'], c='blue', label='Normal Points')
plt.scatter(anomalies_df['Fare_Amount'], anomalies_df['Trip_Distance'], c='red', label='Anomalies')
plt.title('Anomaly Detection using Isolation Forest')
plt.xlabel('Fare Amount')
plt.ylabel('Trip Distance')
plt.legend()
plt.show()
```

```python
#8.Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load dataset from attached CSV file
df = pd.read_csv('AirPassengers.csv', parse_dates=['Month'], index_col='Month')

# Rename column for clarity
df.rename(columns={'#Passengers': 'Passengers'}, inplace=True)

# Define SARIMA parameters
# p, d, q = (1, 1, 1) are orders for AR, differencing, MA parts
# P, D, Q = (1, 1, 1) for seasonal AR, differencing, MA parts with season length s=12 (monthly)
order = (1, 1, 1)
seasonal_order = (1, 1, 1, 12)

# Fit SARIMA model
model = SARIMAX(df['Passengers'], order=order, seasonal_order=seasonal_order)
results = model.fit(disp=False)

# Forecast the next 12 months
forecast_steps = 12
forecast = results.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(df.index[-1] + pd.offsets.MonthBegin(),
                               periods=forecast_steps, freq='MS')

# Extract predicted mean and confidence intervals
forecast_mean = forecast.predicted_mean
forecast_conf_int = forecast.conf_int()

# Plot actual values and forecast
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Passengers'], label='Observed')
plt.plot(forecast_index, forecast_mean, label='Forecast', color='red')
plt.fill_between(forecast_index,
                 forecast_conf_int.iloc[:, 0],
                 forecast_conf_int.iloc[:, 1],
                 color='pink', alpha=0.3)
plt.title('SARIMA Forecast - Monthly Air Passengers')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

```python
#9.Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Generate synthetic 2D numerical dataset (could be replaced by any numerical dataset)
np.random.seed(42)
# Normal points around (50, 10)
X_normal = np.random.randn(300, 2) * [10, 5] + [50, 10]
# Anomalous points far from normal cluster
X_anomalies = np.random.uniform(low=[80, 20], high=[120, 60], size=(20, 2))

# Combine dataset
X = np.vstack([X_normal, X_anomalies])
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])

# Apply LOF for anomaly detection
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.06)
y_pred = lof.fit_predict(X)
df['anomaly'] = y_pred

# Separate normal and anomaly points
normal_points = df[df['anomaly'] == 1]
anomaly_points = df[df['anomaly'] == -1]

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(normal_points['Feature1'], normal_points['Feature2'], c='blue', label='Normal')
plt.scatter(anomaly_points['Feature1'], anomaly_points['Feature2'], c='red', label='Anomaly')
plt.title('Local Outlier Factor (LOF) Anomaly Detection')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
